## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos las siguientes ecuaciones lineales:

* **De Celsius a Fahrenheit:** $$F = (C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de $9/5$)*

* **De Fahrenheit a Celsius:** $$C = (F - 32) \times \frac{5}{9}$$

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir $25$ °C a °F (Temperatura ambiente)
1.  **Identificar la fórmula:** $F = (C \times 1.8) + 32$.
2.  **Sustituir el valor:** $F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar 32:** $45 + 32 = 77$.
* **Resultado:** $25$ °C equivalen a **$77$ °F**.

### Ejemplo 2: Convertir $98.6$ °F a °C (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $C = (F - 32) \div 1.8$.
2.  **Sustituir el valor:** $C = (98.6 - 32) \div 1.8$.
3.  **Restar 32 primero:** $98.6 - 32 = 66.6$.
4.  **Dividir por 1.8:** $66.6 \div 1.8 = 37$.
* **Resultado:** $98.6$ °F equivalen a **$37$ °C**.

## 3. Código en Python

Aquí tienes el script interactivo siguiendo exactamente tu estructura solicitada:

```python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== MENÚ DE CONVERSIONES ===")
        print("1. Convertir grados Centígrados a grados Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")
        
        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```

### Notas sobre el código:
* **`while True`:** Mantiene el programa corriendo hasta que tú decidas cerrarlo.
* **`float()`:** Permite que el usuario ingrese números decimales (ej. 36.5).
* **`f-strings`:** Usé `{valor:.2f}` para que los resultados solo muestren 2 decimales y se vea más limpio.
* **`break`:** Rompe el ciclo infinito cuando eliges la opción 3.

````python
def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    celsius = float(input("\nIngresa los grados Celsius: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngresa los grados Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== MENÚ DE CONVERSIONES ===")
        print("1. Convertir grados Centígrados a grados Fahrenheit")
        print("2. Convertir grados Fahrenheit a grados Centígrados")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Alonso Lopez Garcia NL = 31")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
````
